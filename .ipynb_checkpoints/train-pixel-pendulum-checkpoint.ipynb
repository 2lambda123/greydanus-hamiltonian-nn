```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from IPython import display

from data_pixels import get_dataset
from nn_models import MLPAutoencoder, MLP
from hnn import HNN, HNNBaseline, PixelHNN
from utils import make_gif, L2_loss
```

## Make or load a dataset of pixel-space observations

```python
save_dir = './saved'
name = 'pendulum'
data = get_dataset(name, save_dir, verbose=False)
```

## Inspect the dataset

```python
frames = data['pixels'].reshape(-1,28,28)[:400]

gifname = make_gif(frames, save_dir, duration=1e-1)
display.Image(filename=gifname, width=200)
```

## Make an autoencoder + HNN model

```python
# hypers
side = data['meta']['side']
input_dim = side**2
hidden_dim = 200
latent_dim = 2
learn_rate = 1e-3
batch_size = 200
total_steps = 3000
nonlinearity = 'tanh'
seed = 0

torch.manual_seed(seed)
np.random.seed(seed)

# init model and optimizer
autoencoder = MLPAutoencoder(input_dim, hidden_dim, latent_dim, nonlinearity='relu')
model = PixelHNN(latent_dim, hidden_dim, autoencoder, nonlinearity=nonlinearity, baseline=False)
optim = torch.optim.Adam(model.parameters(), learn_rate)
```

## Train it

```python
# arrange data
trials = data['meta']['trials']
timesteps = data['meta']['timesteps']
inputs = torch.tensor( data['pixels'], dtype=torch.float32)
inputs_next = torch.tensor( data['next_pixels'], dtype=torch.float32)

# vanilla ae train loop
for step in range(total_steps+1):
    ixs = torch.randperm(trials*timesteps-2*trials)[:batch_size] # select a batch

    # get data
    x = inputs[ixs]
    x_next = inputs_next[ixs]

    # loss for training the hnn
    z = model.encode(x)
    z_next = model.encode(x_next)
    z_hat_next = z + model.time_derivative(z)
    hnn_loss = L2_loss(z_next, z_hat_next)

    # loss for canonical coordinates
    w, dw = z.split(1,1)
    w_next, _ = z_next.split(1,1)
    cc_loss = L2_loss(dw, w_next - w)

    # loss for training autoencoder
    x_hat = model.decode(z)
    ae_loss = L2_loss(x, x_hat)

    # combine losses and get gradients
    loss = cc_loss + ae_loss + 1e-2 * hnn_loss
    loss.backward() ; optim.step() ; optim.zero_grad()

    if step % 250 == 0:
        print("step {}, loss {:.4e}".format(step, loss.item()))
```

```python
# step 0, loss 2.8625e-01
# step 250, loss 3.8709e-03
# step 500, loss 1.8506e-03
# step 750, loss 1.1626e-03
# step 1000, loss 7.2703e-04
# step 1250, loss 4.8976e-04
# step 1500, loss 3.0557e-04
# step 1750, loss 2.1330e-04
# step 2000, loss 1.7318e-04
# step 2250, loss 1.4400e-04
# step 2500, loss 1.1108e-04
# step 2750, loss 8.7412e-05
# step 3000, loss 7.0537e-05
```

## Integrate HNN in latent space

```python
z1 = .5
z2 = 0
np_init_z = np.array([[z1, z2]])

contour_steps = 1200
energize_steps = 50
lr = 4e-2

z = torch.tensor( np_init_z, requires_grad=True, dtype=torch.float32)
fut_coords, fut_energies, fut_frames = [], [], []
for i in range(contour_steps + energize_steps + 2*contour_steps):
    np_z = z.detach().numpy()
    np_e = model.hnn(z)[1].item() # first energy
    np_f = model.decode(z).detach().numpy().reshape(side,side)
    if i % 200 == 0:
        print("step {}: coords: ({:.2f}, {:.2f}), energy: {:.3e}".format(i, np_z[0,0], np_z[0,1], np_e))

    fut_coords.append(np_z)
    fut_energies.append(np_e)
    fut_frames.append(np_f.reshape(28,28))
  
    if i < contour_steps or i > contour_steps + energize_steps:
        dz = model.hnn.time_derivative(z)
        # integrate along energy contour
        z = torch.tensor( np_z + lr*dz.detach().numpy(), requires_grad=True, dtype=torch.float32)
    else:
        _, F2 = model.hnn(z)
        dz = torch.autograd.grad(F2.sum(), z, create_graph=True)[0]
        # integrate towards higher energy
        z = torch.tensor( np_z + lr*dz.detach().numpy(), requires_grad=True, dtype=torch.float32)
    
fut_coords = np.concatenate(fut_coords).T
```

```python
# set domain
gs = 60
xmin, xmax = G_[:,0].min(), G_[:,0].max()
vmin, vmax = G_[:,1].min(), G_[:,1].max()
f = 0
xmin -= f
xmax += f
vmin -= f
vmax += f

gX, gV = np.meshgrid(np.linspace(xmin, xmax, gs), np.linspace(vmin, vmax, gs))
np_mesh_inputs = np.stack([gX.flatten(), gV.flatten()]).T
mesh_inputs = torch.tensor( np_mesh_inputs, requires_grad=True, dtype=torch.float32)

# get scalar fields
F1, F2 = model.hnn(mesh_inputs)
np_F1, np_F2 = F1.data.numpy().reshape(gs,gs), F2.data.numpy().reshape(gs,gs)

# plot
fig = plt.figure(figsize=(4, 3), facecolor='white', dpi=200)
ax = fig.add_subplot(1, 1, 1, frameon=True)
plt.contourf(gX, gV, np_F2, cmap='gray_r', levels=20)
ax.set_xlabel("$\\theta$") ; ax.set_ylabel("$\dot \\theta$")
plt.title("Pendulum phase space")

m = 20; k = 3550 ; t = 30
c = np.linspace(0, 10, int(k/m)-1)
plt.scatter(fut_coords[0,t:k][::m], fut_coords[1,t:k][::m], c=c, cmap='viridis', label='data points')
plt.scatter(np_init_z[0,0], np_init_z[0,1], c='red', label='init location')

plt.legend(fancybox=True, framealpha=1.0)
plt.show()
```
